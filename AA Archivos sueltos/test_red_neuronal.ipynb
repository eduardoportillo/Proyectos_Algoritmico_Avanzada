```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_circles
```

```python
df = pd.read_csv('test.csv')
```

```python
df.Gender= df.Gender.map({'Female': 0, 'Male': 1})
df['Customer Type']=df['Customer Type'].map({'Loyal Customer': 0, 'disloyal Customer':1})
df['Type of Travel']=df['Type of Travel'].map({'Business travel': 0, 'Personal Travel':1})
df.Class=df.Class.map({'Eco': 0, 'Business':1,'No servicio tel':1})
df.satisfaction=df.satisfaction.map({'satisfied': 0, 'neutral or dissatisfied':1})
```

```python
df=df.drop('Unnamed: 0', axis=1)
df=df.drop('id', axis=1)
```

```python
df.fillna(df.mean())
df=df.fillna(df.mean(),axis=0)
```

```python
df.describe()
```

```python
df
```

```python
train_x = df.drop('Class', axis=1)
train_y = df['Class']
```

```python
train_y
```

```python
classifier = MLPClassifier(hidden_layer_sizes=(10,5), activation='logistic', solver='sgd', learning_rate='adaptive', learning_rate_init=0.05)
```

```python
classifier.fit(train_x, train_y)
```

```python
print(f"Coeficiente de PÃ©rdida: {classifier.loss_}")
print(f"PrecisiÃ³n de la Red: {classifier.score(train_x, train_y)}")
print(f"NÃºmero de Ã©pocas de estudio: {classifier.n_iter_}")
```
