```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('acquisitions.csv')
```

## Iniciamos Limpieza.

```python
df=df.drop(['id', 'acquisition_id','acquiring_object_id', 'acquired_object_id', 'acquired_at', 'source_url', 'source_description', 'created_at', 'updated_at', 'price_currency_code'],axis=1)
```

```python
df = df.dropna(subset=['term_code'])
```

```python
df.term_code= df.term_code.map({'cash': 0, 'stock': 1, 'cash_and_stock':2})
```

## Iniciamos DBSCAN

```python
clusters = DBSCAN(eps=2, min_samples=10).fit_predict(df)
```

## Graficamos

```python
plt.figure(figsize=(7.5,7.5))
plt.scatter(df.price_amount, df.term_code, c=clusters, s=100, cmap='viridis')
plt.xlabel('price_amount')
plt.ylabel('term_code')
plt.box(False)
plt.show()
```
